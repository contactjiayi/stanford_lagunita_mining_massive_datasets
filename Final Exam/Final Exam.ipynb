Suppose there are six items A, B, C, D, E, and F. In a run of Toivonen's Algorithm, we find that in the sample, the following sets are the maximal frequent itemsets (that is, they and all their subsets are frequent, while no other itemsets are frequent in the sample): ABC, AD, BD, CD, and E.

Which of the following is NOT in the negative border?

```python
from itertools import combinations

items = [letter for letter in 'ABCDEF']
max_freq_itemsets = [['A', 'B', 'C'],
                     ['A', 'D'],
                     ['B', 'D'],
                     ['C', 'D'],
                     ['E']]

def all_freq_itemsets(max_freq_itemsets):
    freq_itemsets = max_freq_itemsets 
    for itemset in max_freq_itemsets:
        for i in range(1, len(itemset)):
            combos = combinations(itemset, i)
            for combo in combos:
                if list(combo) not in freq_itemsets:
                    freq_itemsets.append(list(combo))
    return freq_itemsets

def in_negative_border(candidate, freq_itemsets):
    n_minus_1_subsets = combinations(candidate, len(candidate) - 1)
    if candidate in freq_itemsets:
        return False
    for subset in n_minus_1_subsets:
        if list(subset) not in freq_itemsets:
            return False
    return True

freq_itemsets = all_freq_itemsets(max_freq_itemsets)

for itemset in max_freq_itemsets:
    last_item = sorted(itemset)[-1]
    for idx, item in enumerate(items):
        if idx > items.index(last_item) and item not in itemset:
            candidate = sorted(itemset + [item])
            if in_negative_border(candidate, freq_itemsets):
                print(candidate)
```

Suppose we are executing the DGIM algorithm for approximate counting of bits in a stream. Let the window size be 1000.

(a) What is the largest possible bucket size in the representation of this stream?

(b) Suppose each of the last 1000 bits is 1. What is the smallest possible size of the largest bucket in the representation of the current window?

```python
WINDOW_SIZE = 1000
MAX_NUM_BUCKETS_FOR_SIZE = 2

total_bits = 0
power = 0

while total_bits < WINDOW_SIZE:
    count = 0
    bucket_size = 2 ** power
    while count < MAX_NUM_BUCKETS_FOR_SIZE:
        total_bits += bucket_size
        count += 1
    power += 1

print('Largest possible bucket size: ', 2 ** np.floor(np.log2(1000)))
print('Largest bucket has size ', bucket_size)
```

What estimate does h give as the number of distinct elements in this stream?

```python
stream = []
for n in range(1, 11):
    stream.extend([n] * n)

def int_to_32bit(n):
    return '{:032b}'.format(n)

def num_trailing_zeroes(bits):
    return len(bits) - len(bits.rstrip('0'))

def max_trailing_zeroes(bit_strings):
    return max([num_trailing_zeroes(bs) for bs in bit_strings])

# The 0th moment is the number of distinct elements, i.e. 10
print('The 0th moment is 10')

# The Flajolet-Martin algorithm estimate of the number of distinct elements
# is 2 to the power of the largest tail length of any element in the stream
print('The estimate of the number of distinct elements is {}'.format(
    2 ** max_trailing_zeroes([
        int_to_32bit(n)
        for n in range(1, 11)
    ]),
))

# The 2nd moment or surprise number is the sum of the squares of the
# number of occurrences of each distinct element in the stream
print('The 2nd moment is {}'.format(
    sum([i**2 for i in range(1, 11)])
))

# The expected value of the Alon-Matias-Szegedy algorithm for any variable
# is the 2nd moment of the stream from which it is constructed
print('The AMS estimate of the 2nd moment is the actual 2nd moment \
since we take each position of the stream as a variable.')
```

(c) Suppose we instead construct a support-vector machine to separate these points, using the positive point (2,3) and the negative points (0,2) and (2,4) as the support vectors. The separating hyperplane is defined by equation w.x+b=0. We want to maximize the margin, which means minimizing the length of vector w subject to the constraints that w.x+b>+1 for all positive points x and w.x+b<-1 for all negative points x. What is the sum of the components of vector w?

(d) What is the value of b?

```python
positive_points = np.array([[2, 3],
                            [1, 1]])

negative_points = np.array([[2, 4],
                            [0, 2]])

# The lowest possible slope is the slope of (0, 2) and (2, 3),
# i.e. (3 - 2) / (2 - 0) = 1/2

# The highest possible slope is the slope of (2, 4) and (1, 1),
# i.e (4 - 1) / (2 - 1) = 3

# gradient = 1, intercept = 2
# v = 1 * u + 2
grad, intercept = np.polyfit(
    [pt[0] for pt in negative_points],
    [pt[1] for pt in negative_points],
    deg=1,
)
print(grad, intercept)

# (2, 3) is the positive point closest to the line made by the negative points
# c = y - mx, i.e 3 - 1 * 2 = 1
x, y = (2, 3)
pos_intercept = y - grad * x
print(pos_intercept)

# the parallel line between these 2 lines has the average intercept,
# which turns out to be 3/2, or 1.5
avg_intercept = np.mean([intercept, pos_intercept])

# therefore the equation of the parallel line is
# v = 1 * u + 3/2
w = [-1, 1]
b = -3/2
scaling_factor = 1 / (w @ positive_points[0] + b)
print(scaling_factor)
scaled_w = w * np.array(scaling_factor)
print(scaled_w)
```
